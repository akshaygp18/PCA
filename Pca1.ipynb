#### Steps involved in executing the PCA.
-------------------------------------------------------------------------------------------------------------------------- 
##### 1) First step is to scale the data that means the mean should be equal to zero and standard deviation should be equal to one.
##### 2) Find out the covariance matrix.
##### 3) Find out the respective eigen values and eigen vectors.
##### 4) The eigen value which is having highest value take that respective eigen vectors and do the matrix multiplication with the dataset, that is PCA1.
##### 5) To find out PCA2 consider the second highest eigen value and the repective eigen vectors and do the matrix multiplication of that eigen vectors with dataset.
##### 6) check the model's explained variance ratio values.
##### 7) Plot the screenplot that is nothing but EVR(Explained Varaince Ratio) VS number of columns.
##### 8) Then consider how many columns to select from the screenplot.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_scaled = scaler.fit_transform(df)
```

```python
df1 = pd.DataFrame(df_scaled)
```

```python
df1.describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(df1)
```

```python
df1.shape
```

```python
pd.DataFrame(pca.fit_transform(df1))
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of columns')
plt.ylabel('EVR')
```

##### From the above graph we can conclude that, we can select 5 or 6 number of columns instead of 9
##### so let's try with considering number of columns = 5

```python
pca1 = PCA(n_components=5)
```

```python
new_data = pca1.fit_transform(df1)
```

```python
x = pd.DataFrame(new_data,columns=['PC1','PC2','PC3','PC4','PC5'])
```
